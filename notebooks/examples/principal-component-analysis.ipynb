```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
X = np.array([[-1,-1,-1], 
              [-2,-1,-1], 
              [-3,-2,-1], 
              [1,1,1], 
              [2,1,1], 
              [3,2,1]])
pca = PCA(n_components=2)
pca.fit(X)
```

```python
pca.explained_variance_ratio_
```

```python
pca.singular_values_
```

```python
pca = PCA(n_components=2, svd_solver='full')
pca.fit(X)
```

```python
pca.explained_variance_ratio_
```

```python
pca.singular_values_
```

```python
pca = PCA(n_components=2, svd_solver='arpack') 
# note with arpack n_components must be less than n_features
pca.fit(X)
```

```python
pca.explained_variance_ratio_
```

```python
pca.singular_values_
```
